# combination of the two dataSets

```python
# this are the importand includes used 

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
normalData = pd.read_csv('dataSet/finalData.csv', index_col=0)
attackData = pd.read_csv('dataSet/finalData_attack.csv', index_col=0)

normalData.head()
```

```python
attackData.head()
```

```python
#### now we combine the data sets randomly
```

```python
dataCom = [normalData, attackData]

combined_dataset = pd.concat(dataCom)
combined_dataset
```

```python
newSet = combined_dataset
```

```python
shuffledSet = newSet.sample(frac=1)
```

```python
shuffledSet
```

```python
finalDataSet = shuffledSet.reindex()
```

```python
finalDataSet
```

```python
plt.plot(finalDataSet)
```

```python
finalDataSet.to_csv(f"dataSet/combinedData.csv")
```
